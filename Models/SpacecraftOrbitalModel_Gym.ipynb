```python
# Register the custom environment with Gym
gym.register(
    id="SpacecraftOrbitalEnv-v0",
    entry_point=lambda config: env_creator(config, "orbital")
)
# Verify environment registry
gym.pprint_registry()
```

```python
# Make our environment
gym_env = gym.make("SpacecraftOrbitalEnv-v0", config=config)
```

```python
# Initialize the PPO agent
# stable_baselines3

# Wrap the environment for vectorized training
vec_env = make_vec_env(lambda: gym_env, n_envs=1)

model = PPO(
    "MlpPolicy",
    gym_env,
    verbose=1,
    learning_rate=3e-4,  # Default learning rate (tunable)
    batch_size=256,  # Larger batch size helps with training stability
    gamma=0.99,  # Discount factor
)
```

```python
# Train the agent
# using PPO and SAC algorithms for this application
model.learn(total_timesteps=1_000_000)

# Save the trained model
model.save("spacecraft_ppo")
```

```python
# Evaluate the agent
# Load agent
model = PPO.load("spacecraft_ppo")
obs, _ = gym_env.reset()
done = False
```

```python
while not done:
    action, _states = model.predict(obs)
    obs, reward, done, _, _ = gym_env.step(action)
    print(f"Reward: {reward}")
```
